```python
%matplotlib inline
import math
import random
import numpy as np
import pickle
import matplotlib.pyplot as plt
import scipy.io as sio
```

```python
file_load1 = open('./HCWS/Record_moving1.pkl', 'rb')
Record_load1 = pickle.load(file_load1)
Record1 = np.asarray(Record_load1[::1])
w1 = Record1[:,2]
maxw1 = Record1[:,3]

file_load2 = open('./HCWS/Record_moving2.pkl', 'rb')
Record_load2 = pickle.load(file_load2)
Record2 = np.asarray(Record_load2[::1])
w2 = Record2[:,2]
maxw2 = Record2[:,3]

file_load3 = open('./SKY/Record_moving1.pkl', 'rb')
Record_load3 = pickle.load(file_load3)
Record3 = np.asarray(Record_load3[::1])
w3 = Record3[:,2]
maxw3 = Record3[:,3]

file_load4 = open('./SKY/Record_moving2.pkl', 'rb')
Record_load4 = pickle.load(file_load4)
Record4 = np.asarray(Record_load4[::1])
w4 = Record4[:,2]
maxw4 = Record4[:,3]
```

```python
x = np.arange(0, len(Record1), 1)

plt.figure(0)
plt.plot(x, w1)
plt.plot(x, w2)
plt.plot(x, w3)
plt.plot(x, w4)
plt.title("w")

plt.figure(1)
plt.plot(x, maxw1)
plt.plot(x, maxw2)
plt.plot(x, maxw3)
plt.plot(x, maxw4)
plt.title("maxw")

sio.savemat('record.mat', 
            {'x':x,
             'w1':w1, 'w2':w2, 'w3':w3, 'w4':w4,
             'maxw1':maxw1, 'maxw2':maxw2, 'maxw3':maxw3, 'maxw4':maxw4})
```
